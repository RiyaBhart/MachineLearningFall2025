```python
import pandas as pd

dfA = pd.read_csv(r'/content/Lab2 D1A.csv')
dfB =  pd.read_csv(r'/content/Lab2 D1B.csv')
dfC = pd.read_csv(r'/content/Lab2 D1C.csv')

print(dfA.shape)
print(dfB.shape)
print(dfC.shape)

print(dfA.columns)
print(dfB.columns)
print(dfC.columns)
```

```python
common_cols = dfA.columns.intersection(dfB.columns)
print("Duplicate columns:", common_cols.tolist())
```

```python
dfB_unique = dfB.drop(columns=common_cols)
combined_AB = pd.concat([dfA, dfB_unique], axis=1)
print(combined_AB.shape)
print(combined_AB.columns)
```

```python
common_keys = dfA.columns.intersection(dfC.columns)
print("Possible merge keys:", common_keys.tolist())
```

```python
comboAC = pd.merge(dfA, dfC, on='county', how='inner')
print(comboAC.shape)
print(comboAC.columns)
```

```python
from imblearn.over_sampling import SMOTE

df = df.dropna(subset=['TARGET'])

X = df.drop('TARGET', axis=1)
y = df['TARGET']

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

print(y_resampled.value_counts())
```

```python
corr_target = df.corr()['TARGET'].abs()
selected_features = corr_target[corr_target > 0.65]
print("Selected features:", selected_features)
```
