```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data Preprocessing
# A. downloaded the dataset
# B. importing dataset
df = pd.read_csv(r'/content/cancer patient data sets.csv')
```

```python
# C. Unique category in targer variable
df['Level'].unique()
```

```python
# D. Number of columns in dataset, 50 tops rows and 30 bottom rows
total_columns = df.shape[1]
print("total columns in dataset : ",total_columns)

top_rows = df.head(50)
print(top_rows)

bottom_rows = df.tail(30)
print(bottom_rows)

print(df.info())

print("shape : ",df.shape)

print("memory usage : ",df.memory_usage().sum())
```

```python
# E. number of columns and samples
print("number of Columns:", df.shape[1])
print("number of Samples:", df.shape[0])
```

```python
# F. number of records for each unique category of target variable using numerical and graphical visualization.
print(df['Level'].value_counts())
df['Level'].value_counts().plot(kind='bar',color=['red','green','blue'])
plt.show()
```

```python
# G. extract samples using loc (row 550 to 900, columns Gender to Coughing of Blood)
subset1 = df.loc[550:900, "Gender":"Coughing of Blood"]

# save new dataset
subset1.to_csv("subset1.csv", index=False)

# statistical analysis
print("Mean:\n", subset1.mean())
print("\nMedian:\n", subset1.median())
print("\nMode:\n", subset1.mode().iloc[0])
print("\nStandard Deviation:\n", subset1.std())
print("\nMinimum:\n", subset1.min())
print("\nMaximum:\n", subset1.max())
```

```python
# H. extract last 20 rows and 5 features (Shortness of Breath to Frequent Cold)
subset2 = df.iloc[-20:, df.columns.get_loc("Shortness of Breath"):df.columns.get_loc("Frequent Cold")+1]

# save new dataset
subset2.to_csv("subset2.csv", index=False)

# statistical analysis
print("Mean:\n", subset2.mean())
print("\nMedian:\n", subset2.median())
print("\nMode:\n", subset2.mode().iloc[0])
print("\nStandard Deviation:\n", subset2.std())
print("\nMinimum:\n", subset2.min())
print("\nMaximum:\n", subset2.max())
```

```python
# I. Extract these rows
records = df.loc[[110, 220, 360, 440, 656, 778, 202, 889], :]

# Average alcohol use
print("Average Alcohol Use:", records["Alcohol use"].mean())
```

```python
# A. Graphical visualization

# Target variable distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Level', data=df, palette="pastel")
plt.title("Target Variable Distribution", fontsize=14)
plt.xlabel("Level")
plt.ylabel("Count")
plt.show()

# Pairplot for quick visualization of relationships
sns.pairplot(df[["Age", "Obesity", "Frequent Cold"]], diag_kind="kde")
plt.suptitle("Pairplot of Age, Obesity & Frequent Cold", y=1.02, fontsize=14)
plt.show()
```

```python
# Correlation graph
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".1f")
plt.title("Correlation Heatmap", fontsize=14)
plt.show()
```

```python
# Distribution of Age
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], kde=True, bins=20, color="blue")
plt.title("Distribution of Age", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Distribution of Frequent Cold
plt.figure(figsize=(8,5))
sns.countplot(x="Frequent Cold", data=df, palette="viridis")
plt.title("Distribution of Frequent Cold", fontsize=14)
plt.xlabel("Frequent Cold")
plt.ylabel("Count")
plt.show()
```

```python
# Outliers using Boxplot
plt.figure(figsize=(8,5))
sns.boxplot(x=df["Obesity"], color="orange")
plt.title("Outliers in Obesity Feature", fontsize=14)
plt.show()

# Outliers using Violin plot (alternative view)
plt.figure(figsize=(8,5))
sns.violinplot(x=df["Obesity"], color="lightgreen")
plt.title("Violin Plot for Obesity Feature", fontsize=14)
plt.show()
```
